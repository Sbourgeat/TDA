```python
import numpy as np
import igraph
```

```python
A = np.load('im.tiff_adj_matrix.npy', allow_pickle=True)
```

```python
A = A.tolist()
```

```python
A2 =[]
for i in A:
    A2.append(i)
```

```python
g = igraph.Graph.Adjacency(A2, directed=False)
g = g.as_undirected()
```
